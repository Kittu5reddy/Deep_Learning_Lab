**Implementation of M-L-P for CIFAR10 Dataset**

import packages

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.losses import categorical_crossentropy
from keras.callbacks import CSVLogger

csv_logger = CSVLogger('./logs/cifar10_training_log.csv', append=True)
```

load dataset

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

check the data

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
```

convert y_test and y_train into one hot encoding

```python
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
y_train
```

```python
y_test
```

create the nueral architech

```python
model=Sequential()
layer_one=Flatten(input_shape=[32,32,3])
output_layer=Dense(10,activation='softmax')

```

```python
model.add(layer_one)
model.add(output_layer)
```

1.training model with optm-adam loss=categorical_crossentropy

```python
model.compile(optimizer='adam',loss=categorical_crossentropy)
model.fit(x_test,y_train,epochs=20,batch_size=64,callbacks=[csv_logger])
```

```python
loss,accurasy=model.evaluate(x_test,y_test)
print("loss ",loss)
print("accurasu ",accurasy)
```

2. training with metric(accuracy)

```python
model.compile(optimizer='adam',loss=categorical_crossentropy)
histrory=model.fit(x_test,y_train,epochs=20,batch_size=64,callbacks=[csv_logger])
```

```python
second_loss_test,second_accurcy_test=model.evaluate(x_test,y_test)
print("loss",second_loss_test)
print("accurasy",second_accurcy_test)
```

visualizing  loss

```python
plt.plot(histrory.history['loss'])
plt.title("train_loss")
plt.show()
```

printing the accuracy


```python
plt.plot(histrory.history['accuracy'])
plt.title("train_accuracy")
plt.show()
```
